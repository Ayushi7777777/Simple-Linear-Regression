## Independent and dependent featues

```python
from sklearn.model_selection import train_test_split
```

```python
x = dataset # independent
y= df.target # dependent
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=30, random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
x_train= scaler.fit_transform(x_train)
```

```python
x_test= scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
## cross validation 
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
mean_squrred_error= cross_val_score(regression, x_train , y_train, cv = 20 , scoring='neg_mean_squared_error') #cv = cross validation )
```

```python
np.mean(mean_squrred_error)
```

```python
##prediction
regresson_predict= regression.predict(x_test)
```

```python
regresson_predict
```

```python
sns.displot(regresson_predict- y_test , kind='kde' )
```

```python
from sklearn.metrics import r2_score
```

```python
score= r2_score(regresson_predict , y_test)
```
